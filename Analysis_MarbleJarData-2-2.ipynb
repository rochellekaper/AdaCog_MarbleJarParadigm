```python
# To do stats for jar selection task:

# average utility of their choice for each subject

# for each subject, average all utilities

# Q: On avg, was the utility of choices higher /  lower in stressor group? --> t test
```

```python
def calculate(col):
    '''Input is the Marble Jar column (ex: data['Jar1_Colors']). This function will
    take into account the marble color values and generate a list of the marble jar values.
    Then the sum, mean, and variance will be calculated. Lists of all of the rows' marble jar values, means,
    and variances will be outputted so they can then be added to the dataframe.'''

    #mean_list = []
    var_list = []
    marbles_list = []

    for val, g, o, p in zip(col, data['green_value'], data['orange_value'], data['purple_value']):
        green_val = g
        orange_val = o
        purple_val = p
        green_sum = val.count('green') * g
        orange_sum = val.count('darkorange') * o
        purple_sum = val.count('purple') * p
        total = green_sum + orange_sum + purple_sum

        colors = val
        #Convert colors to its values + get rid of extra characters
        marbles = colors.replace('green', str(g)).replace('darkorange', str(o)).replace('purple', str(p)).replace('.', '').strip()
        marbles = marbles.replace("''", "").replace("'", "").replace("[", "").replace("]", "").replace(",", "").strip()
        marbles = list(marbles.replace('"', ""))

        #remove extra white spaces
        while (" ") in marbles:
            marbles.remove(" ")

        marbles = list(map(int, marbles))
        marbles_list.append(marbles)

        #Calculate mean & variance and create lists so we can add as columns to df
#         mean = total / 9
#         mean_list.append(mean)
        var = statistics.variance(marbles)
        var_list.append(var)

    return var_list, marbles_list

```

```python
## Add columns to df. Ignore the warning
var_list1, marbles_list1 = calculate(data['Jar1_Colors'])
data['Jar1_Variance'] = var_list1
data['Jar1_MarblesVal_List'] = marbles_list1
var_list2, marbles_list2 = calculate(data['Jar2_Colors'])
data['Jar2_Variance'] = var_list2
data['Jar2_MarblesVal_List'] = marbles_list2
var_list3, marbles_list3 = calculate(data['Jar3_Colors'])
data['Jar3_Variance'] = var_list3
data['Jar3_MarblesVal_List'] = marbles_list3
```

```python
# To do stats for jar selection task:

# average utility of their choice for each subject

# for each subject, average all utilities

# Q: On avg, was the utility of choices higher /  lower in stressor group? --> t test
```

```python
def calculate_expected_utility(jar_colors_col, green_val, orange_val, purple_val):
    expected_utility_list = []

    for jar, g, o, p in zip(jar_colors_col, green_val, orange_val, purple_val):

        #convert jar to list
        jar = jar.replace('.', '').strip()
        jar = jar.replace("''", "").replace("'", "").replace("[", "").replace("]", "").replace(",", "").strip()
        jar = jar.replace('"', "")
        jar = jar.split(" ")


        # Define utility function for each color
        utility_mapping = {'green': g, 'darkorange': o, 'purple': p}

        # Count occurrences of each color
        counts = {color: jar.count(color) for color in set(jar)}

        # Calculate probability for each color
        probabilities = {color: count / len(jar) for color, count in counts.items()}

        # Calculate expected utility and add to the list
        expected_utility = sum(probabilities[color] * utility_mapping[color] for color in set(jar))
        expected_utility_list.append(expected_utility)

    return expected_utility_list
```

```python
jar1_utility = calculate_expected_utility(data['Jar1_Colors'], data['green_value'], data['orange_value'], data['purple_value'])
jar2_utility = calculate_expected_utility(data['Jar2_Colors'], data['green_value'], data['orange_value'], data['purple_value'])
jar3_utility = calculate_expected_utility(data['Jar3_Colors'], data['green_value'], data['orange_value'], data['purple_value'])
```

```python
data['jar1_Expected_Utility'] = jar1_utility
data['jar2_Expected_Utility'] = jar2_utility
data['jar3_Expected_Utility'] = jar3_utility
```
